## Initial Data Loading

### Staging Layer

```sql
CREATE TABLE dw_training.staging_sales
AS
Select * FROM invoice;

select * from staging_sales;
```

### Transformation

```sql
-- creating a view which is just a data view and not data

CREATE OR REPLACE VIEW dw_training.staging_sales_view
AS
select orderid, customeremail, regionname, totalamount,
totalamount * 0.9 as discountamount
from
staging_sales
```

```sql
CREATE OR REPLACE VIEW dw_training.national_sales_view
AS
select country,
SUM(totalamount * 0.9) as nationaldiscounttotal,
count(orderid) as totalorders
from
staging_sales
group by country
ORDER BY totalorders asc;


select * from national_sales_view;
```

### Core Layer

```sql
CREATE OR REPLACE TABLE dw_training.sales_core (
    orderid INT NOT NULL,
    customeremail VARCHAR(150) NOT NULL,
    regionname VARCHAR(50) NOT NULL,
    totalamount DECIMAL(12,2) NOT NULL,
    discountamount DECIMAL(12,2) NOT NULL
);

INSERT INTO sales_core
SELECT *
FROM dw_training.staging_sales_view;

SELECT *
FROM sales_core;
```

## Incremental Data Loading

```sql
-- DATA UPSERT

INSERT INTO invoice
SELECT *
FROM invoice_new;

select count(*), min(orderdate), max(orderdate)
from invoice;
```

```sql
-- INCREMENTAL LOAD

CREATE OR REPLACE TABLE dw_training.staging_sales
AS
Select * FROM invoice
Where orderdate > '2024-01-24';

SELECT * FROM dw_training.staging_sales;
```

```sql
-- UPDATE TRANSFORMED DATA VIEW
CREATE OR REPLACE VIEW dw_training.staging_sales_view
AS
select orderid, customeremail, regionname, totalamount,
totalamount * 0.9 as discountamount
from
staging_sales
```

```sql
INSERT INTO sales_core
SELECT *
FROM dw_training.staging_sales_view;

SELECT *
FROM sales_core;
```

# Dimensional Data Modelling
Creating Fact and Dimension Tables


```sql
CREATE OR REPLACE TABLE dw_training.invoice_table
AS
select *
from
invoice;

select * from invoice_table;
```

```sql
select regionid, customerid, orderid,
dense_rank() Over (partition by regionid Order by customerid ASC) as row_rank
from invoice_table
```

### Dimension Tables - Creating Views

`ROW_NUMBER()` assigns a new surrogate key per row

```sql
Create or Replace VIEW DimCustomersView
As
select T.*,
row_number() over (Order by T.customerid ASC) as DimCustomerKey
From
(select distinct customerid ,
customername,
customeremail
from invoice_table) as T;

select * from DimCustomersView;
```

```sql
Create or Replace VIEW DimProductsView
As
select T.*,
row_number() over (Order by T.productid ASC) as DimProductKey
From
(select distinct productid,
productname,
productcategory
from invoice_table) as T;

select * from DimProductsView;
```

```sql
Create or Replace VIEW DimRegionsView
As
select T.*,
row_number() over (Order by T.Regionid) as DimRegionKey
From
(select distinct Regionid,
Regionname,
Country
from invoice_table) as T;

select * from DimRegionsView;
```

```sql
Create or Replace VIEW DimDatesView
As
select T.*,
row_number() over (Order by T.Orderdate) as DimDateKey
From
(select distinct Orderdate
from invoice_table) as T;

select * from DimDatesView;
```

### Dimension Tables - Creating Tables

```sql
-- dimcustomers
CREATE OR REPLACE TABLE dimcustomers
(
  Customerid int,
  Customername string,
  Customeremail string,
  DimCustomerKey int
);

INSERT INTO dimcustomers
SELECT *
FROM DimCustomersView;

-- dimproducts
CREATE OR REPLACE TABLE dimproducts
(
  Productid int,
  Productname string,
  Productcategory string,
  DimProductKey int
);

INSERT INTO dimproducts
SELECT *
FROM DimProductsView;

-- dimregions
CREATE OR REPLACE TABLE dimregions
(
  Regionid int,
  Regionname string,
  Country string,
  DimRegionKey int
);

INSERT INTO dimregions
SELECT *
FROM DimRegionsView;

-- dimdates

CREATE OR REPLACE TABLE dimdates
(
  Orderdate date,
  DimDateKey int
);

INSERT INTO dimdates
SELECT *
FROM DimDatesView;
```

### Fact Table Creation

```sql
Create or replace TABLE Fact_table
(
  OrderID int,
  Quantity decimal,
  Unitprice decimal,
  Totalamount decimal, 
  Dimcustomerkey int,
  Dimregionkey int,
  Dimproductkey int,
  Dimdatekey int
);
```

### Fact Table Data Insertion

```sql
INSERT INTO Fact_table
select F.Orderid, F.quantity, F.unitprice, F.totalamount,
dc.DimCustomerKey, dr.DimRegionKey, dp.DimProductKey, dd.DimDateKey
from invoice_table F
inner join dimcustomers dc
USING (customerid)
inner join dimproducts dp
USING (productid)
inner join dimregions dr
on f.country = dr.country
inner join dimdates dd
on f.orderdate = dd.orderdate;

select *
from Fact_table;
```
